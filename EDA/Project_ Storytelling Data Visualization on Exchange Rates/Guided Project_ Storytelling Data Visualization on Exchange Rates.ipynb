# Storytelling Data Visualization on Exchange Rates

In this project, we will explore a dataset that describes Euro daily exchange rates between 1999 and 2021 and use data visualization to compare the exchange rates of American dollar($) and euros(€). If the exchange rate between euro and the US dollar is 1.2, that means you get 1.2 dollars for 1 euro. Thus, euro has more value than US dollar in that exchange rate.

The dataset is made available on Kaggle by Daria Chemkaeve.

For our data storytelling, we will:

* Use a line plot to show how the euro-dollar exchange rate has chnaged during the 2020 coronavirus pandemic.
* Use a line plot to show how the euro-dollar exchange rate has chnaged during the 2007-2008 financial crisis.

Let's start by importing pandas, reading our data into a dataframe and inspecting the dataframe to understand its structure and some basic facts.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

## Data Cleaning

Before visualizing our data, let's do some basic data cleaning. In the following cell, we will do:

* Rename `[US dollar]` to `US_dollar` and `Period\Unit:` to `Date`.
* Change the `Date` column to a `datetime` object.
* We sort the values by Date in ascending order.
* We reset the index (and drop the initial index).


```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar',
                              'Period\\Unit:': 'Time'},
                     inplace=True)
exchange_rates['Time']=pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time',inplace=True)
```

```python
exchange_rates.head()
```

```python
# rest index
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
#Isolate the Time and the US_dollar columns. Assign them to a different variable named euro_to_dollar.
euro_to_dollar=exchange_rates[['Time','US_dollar']]
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
#Drop all the rows where the - character appears in the US_dollar column.
#method 1
euro_to_dollar.drop(euro_to_dollar.copy()[euro_to_dollar['US_dollar']=='-'].index,inplace=True)
euro_to_dollar['US_dollar'].value_counts()
```

```python
#method 2
euro_to_dollar=exchange_rates[['Time','US_dollar']]
euro_to_dollar=euro_to_dollar[euro_to_dollar['US_dollar']!='-']
euro_to_dollar['US_dollar'].value_counts()
```

```python
#method 3
euro_to_dollar=exchange_rates[['Time','US_dollar']]
euro_to_dollar=euro_to_dollar[~euro_to_dollar['US_dollar'].isin(['-'])]
euro_to_dollar['US_dollar'].value_counts()
```

```python
# convert the US_dollar column to a float data type.
euro_to_dollar['US_dollar']=euro_to_dollar['US_dollar'].astype(float)
#recheck
euro_to_dollar.info()
```

## Rolling Mean

```python
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['US_dollar_mean_30']=euro_to_dollar['US_dollar'].rolling(30).mean()
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar_mean_30'])
plt.title('Rolling Window:30')
plt.show()
```

```python
plt.figure(figsize=(15,12))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title('Orginal Values',weight = 'bold')
for i,roll in zip([2,3,4,5,6],[7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'].rolling(roll).mean())
    plt.title('Rolling Windows:'+str(roll),
         weight='bold')
plt.show()
```

## Storytelling Data Visualization
### 1. Exchange Rate During Coronavirus Pandemic

Unfortunately, the x-coordinates (behind the datetime units) are not fixed in matplotlib (to fix them, you’d need a slightly more complicated coding approach), so the solution here is to use `ax.get_xticks()` and `ax.get_yticks()` to see what are the coordinates and then adjust your input coordinates accordingly.

In this section, we will create a line plot using 'fivethirtyeight' style and other matplotlib functions to tell the story of how euro-dollar rate changes during the coronavirus pandemic of 2020.

Thus, we will focus our attention to only few years before and after the pandemic.

Let's use 2018-2021 as our base and highlight the 2020-2021 part.

```python
# Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
# Create a dataframe with range of 2018 - 2023

baseline = euro_to_dollar[(euro_to_dollar['Time'].dt.year>=2018) &
                          (euro_to_dollar['Time'].dt.year<=2023)]
# Create a dataframe with range representative of pandemic peak

cov19 = euro_to_dollar[(euro_to_dollar['Time'].dt.year>=2020) &
                          (euro_to_dollar['Time'].dt.year<=2022)]
```

```python
#adding plot
fig, ax = plt.subplots(figsize=(11,6))
ax.plot(baseline['Time'],baseline['US_dollar'].rolling(30).mean(),
        linewidth=1, color='#0000FF')
ax.plot(cov19['Time'],cov19['US_dollar'].rolling(30).mean(),
        linewidth=3, color='#FF0000')

# highlighting lowest rate during the crisis
ax.axvspan(xmin=737535.0, xmax=737600.0, ymin=0,
           alpha=0.3, color='grey')

## Adding COVID case info text on plot    
ax.text(737529, 1.121, 'lowest', color='#af0b1e',
         weight='bold', size='small', rotation='horizontal')

## Adding titles
ax.text(736490, 1.26, "Euro-USD rate dipped to 1.085 during 2020 coronavirus pandemic",
        weight='bold', size=20)
ax.text(736490, 1.25, 'Euro-USD exchange rates between 2018 and 2021',
        size=16)

# making the grid lines transparent
ax.grid(alpha=0.3)

plt.show()
```

### 2. Exchange Rate During Financial Crisis

```python
# Create a dataframe with range of 2006-2009

baseline = euro_to_dollar[(euro_to_dollar['Time'].dt.year>=2006) &
                          (euro_to_dollar['Time'].dt.year<=2009)]

# Create a dataframe with range representative of financial crisis
crisis = euro_to_dollar[(euro_to_dollar['Time'].dt.year>=2007) &
                          (euro_to_dollar['Time'].dt.year<=2008)]
```

```python
# adding plot
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(11,6))
ax.plot(baseline['Time'],baseline['US_dollar'].rolling(30).mean(), linewidth=1, color="#A6D785",label = 'Baseline')
ax.plot(crisis['Time'],crisis['US_dollar'].rolling(30).mean(),  linewidth=2, color="#e23d28",label = 'Financial Crisis')

#axis labels and legend
plt.ylabel('Exchange Rate', size = 15, weight = 'bold')
ax.legend(loc='upper center', bbox_to_anchor=(0.2525, 0.9375),
          ncol=2, fancybox=True, shadow=True)

# highlighting highest rate during the crisis
ax.axvspan(xmin=733112, xmax=733301, ymin=0,
           alpha=0.3, color='grey')

# Adding case info text on plot    
ax.text(733152, 1.50, 'highest', color='#af0b1e',
         weight='bold', size='small', rotation='horizontal')

# Adding title
ax.text(732172, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold',size=18)
ax.text(732172, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

# Adding source
ax.text(732172, 1.12, '©DATAQUEST' + ' '*165 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

# making the grid lines transparent
ax.grid(alpha=0.3)

plt.show()
```
